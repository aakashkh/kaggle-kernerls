## EDA 

### Passholder Type : 
The name of the passholder's plan.
The following table shows the popularity of different Passholder plans, Monthly Pass and Walk-up both combined are more tha 90%, in which monthly once constitute around just more than 60%. The least poplular according to data is Staff Annual Plan.

```python
kable(round(prop.table(table(bike_data$`Passholder Type`))*100,2))
```

### Plan Duration:
The number of days that the plan the passholder is using entitles them to ride; 0 is used for a single ride plan (Walk-up). The most popular plan duration is monthly, followed by Single Ride  and then yearly. We can covert the factor levels to more descriptive one namely, 30 with Monthly, 365 with Yearly and rest as Single Ride.  

```python
bike_data$`Plan Duration` = as.factor(bike_data$`Plan Duration`)
levels(bike_data$`Plan Duration`) = c('Single Ride','Monthly','Yearly')
kable(round(prop.table(table(bike_data$`Plan Duration`))*100,2))
```

### Trip Route Category :
"Round Trip" for trips starting and ending at the same station or "One Way" for all other trips. One way trip clearly rules with more than 90% users preferring it.

```python
kable(round(prop.table(table(bike_data$`Trip Route Category`))*100,2))
```

### Duration:
 Length of trip is in seconds. Let us create a new column which convert Duration from seconds to minutes (Duration_Mins) for better understanding.

```python
bike_data$Duration_Mins = bike_data$Duration/60
```

We see some blank rows for the column "Plan Duration" of the datasets. Let`s dig a bit more and pull some other columns too.

```python
unique(bike_data$`Plan Duration`)
```

```python
bike_data_subset = subset(bike_data, is.na(bike_data$`Plan Duration`))
kable(table(bike_data_subset$`Plan Duration`,bike_data_subset$`Passholder Type`))
```

We noticed that for Monthly Pass and Staff Annual there are cases where Plan Duration is NA and hence by common understanding, replace all Plan Duration where Passholder Type is "Monthly Pass" by "Monthly" and "Staff Annual" by "Yearly"

```python
bike_data$`Plan Duration`[bike_data$`Passholder Type` == "Monthly Pass"] <- "Monthly" 
bike_data$`Plan Duration`[bike_data$`Passholder Type` == "Staff Annual"] <- "Yearly" 
```

 Group the Dataset by "Trip Route Category","Passholder Type","Plan Duration" and lets calculate the average time of the bike ride for the each combination obtained

```python
bike_data_group <- bike_data %>%
                  group_by(`Trip Route Category`,`Passholder Type`,`Plan Duration`) %>%
                  summarize(average_duration = round(mean(Duration_Mins)))
```

Above code genrates the required table

```python
kable(bike_data_group)
```

 We can plot this over a Bar Graph using ggplot library.

```python
options(repr.plot.width=10, repr.plot.height=6)
ggplot(bike_data_group, aes(x = `Passholder Type`, y = average_duration, fill = `Plan Duration`, label = round(average_duration))) +
  geom_bar(stat = "identity") + facet_grid(~ `Trip Route Category`)+
  geom_text(size = 4.5, position = position_stack(vjust = 0.5)) +
  labs(x = "Passholder Type", y = "Average Duration (Mins)") 
```

## Histogram

Now we`ll create the frequency distribution for no. of Starting Station ID, Ending Staion ID and Bike ID by group them in bands of avergae time spent for the trips associated with them.

### We`ll start with Starting Station ID.

 Step 1: Group the data on Starting station ID and calculate average time associated with them  
 Step 2: Create bands starting from 10 to 500 ( this can be verified with min and max average time of the starting station Ids.) and label them accordingly.  
 Step 3: Group the starting station Ids according to above bins created  
 Step 4: Combine the bins with orignal grouped data and plot using ggplot2 library  

 We will repeat the same step for Ending station Id and Bike Ride.

 Following is the code for the above mentioned steps and below is shown the output in the tabular as well as graphical mode.

```python
start_station_group <- bike_data %>%
  group_by(`Starting Station ID`) %>%
  summarize(average_duration = mean(Duration_Mins))
break_start_station <- c(10,15,20,25,30,35,40,45,50,500)
lab <- c("10-15)", "15-20)", "20-25)", "25-30)", "30-35)", "35-40)", "40-45", "45-50", ">50")
bins_start_station = cut(start_station_group$average_duration, break_start_station,include.lowest = T, right=FALSE, labels= lab)
start_station_group <- cbind(start_station_group, bins_start_station)
```

```python
kable(table(bins_start_station))
```

```python
options(repr.plot.width=6, repr.plot.height=4)
ggplot(start_station_group, aes(x=start_station_group$bins_start_station,fill=..count..)) + 
  geom_bar(color='black', alpha=0.5) + 
  stat_count(geom="text", aes(label=..count..), hjust=-1) +
  theme_bw() + 
  labs(y="No. of Start Station",x="Average Duration (Mins)") +
  coord_flip() +
  ylim(0,40)
```

### Repeating the same steps for Ending Station ID, results into - 

```python
end_station_group <- bike_data %>%
  group_by(`Ending Station ID`) %>%
  summarize(average_duration = mean(Duration_Mins))

break_end_station <- c(10,15,20,25,30,35,40,45,50,500)
bins_end_station = cut(end_station_group$average_duration, break_end_station,include.lowest = T, right=FALSE,labels= lab)
end_station_group <- cbind(end_station_group, bins_end_station)
 
```

```python
kable(table(bins_end_station))
```

```python
ggplot(end_station_group, aes(x=end_station_group$bins_end_station,fill=..count..)) + 
    geom_bar(color='black', alpha=0.5) + 
    stat_count(geom="text", aes(label=..count..), hjust=-1) +
    theme_bw() + 
    labs(y="No of End Station",x="Average Duration (Mins)") +
    coord_flip() +
    ylim(0,40) 
```

### Doing the same thing for Bike ID - 

```python
bikeID_group <- bike_data %>%
  group_by(`Bike ID`) %>%
  summarize(average_duration = mean(Duration_Mins))

break_bikeID <- c(10,15,20,25,30,35,40,45,50,500)
bins_bikeID = cut(bikeID_group$average_duration, break_bikeID,include.lowest = T, right=FALSE,labels= lab)
bikeID_group <- cbind(bikeID_group, bins_bikeID)
```

```python
kable(table(bins_bikeID))
```

```python
options(repr.plot.width=12, repr.plot.height=8)
ggplot(bikeID_group, aes(x=bikeID_group$bins_bikeID,fill=..count..)) + 
    geom_bar(color='black', alpha=0.5) + 
    stat_count(geom="text", aes(label=..count..), hjust=-0.25) +
    theme_bw() + 
    labs(y="No. of Bikes",x="Average Duration (Mins)") +
    coord_flip() 
```
